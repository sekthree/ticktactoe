```python
import random
from IPython.display import clear_output

def check_board(board):
        
    #Assign the board to all possible win scenarios
    positions = {"top":[''.join(board[7:10]),7,8,9],"mid":[''.join(board[4:7]),4,5,6],"bot":[''.join(board[1:4]),1,2,3]}
    positions["left"]   = [f"{''.join(board[1:9:3])}",1,4,7]
    positions["center"] = [f"{''.join(board[2:10:3])}",2,5,8]
    positions["right"]  = [f"{''.join(board[3:10:3])}",3,6,9]
    positions["lrX"]    = [f"{''.join(board[3:8:2])}",3,5,7]
    positions["rlX"]    = [f"{''.join(board[1:10:4])}",1,5,9]
    
    #Check if any scanrios are winners if so highlight the row in red
    #If board is full with no winners stop game play
    #Enumerate through entire positions other wise " " will satisfy true on first iteration when full
    for i,(key,value) in enumerate(positions.items()):
        if 'XXX' in value[0] or 'OOO' in value[0]:
            for cell in value[1:4]:
                if cell in (1,2,3):
                    att = ["\033[91m\033[1m","\033[0m"]
                else:
                    att = ["\033[91m\033[1m\033[4m","\033[0m"]
                board[cell] = f"{att[0]}{board[cell].upper()}{att[1]}"
            print(f"{value[0]} WINS")
            return False,board
        elif " " not in board and i == 7:
            print("Strange game. It's a TIE! The only winning move is not to play.")
            return False,board
        
    return True,board
    
def display_board(board):
    
    #Before displaying, check for winners
    clear_output()
    check, board = check_board(board)     
    
    #Create a printable version of the board to display
    canvas = "\033[1m\033[4mBoard\t\tMapping\033[0m\n"
    canvas += f"\033[4m{'|'.join(board[7:10])}\033[0m\t\t\033[4m7|8|9\033[0m\n"
    canvas += f"\033[4m{'|'.join(board[4:7])}\033[0m\t\t\033[4m4|5|6\033[0m\n"
    canvas += f"{'|'.join(board[1:4])}\t\t1|2|3\n"
        
    #Print the canvas/board
    print(canvas)
    
    return check
    
def user_move(board):

    #Varibales for displaying for userinput
    message = "Enter a valid number between 1-9: "
    error = ""
    
    while True:
        #Ask for user input and display any previous errors
        move = input(f"{error}{message}")
        error = ""
        
        #check user input for errors
        if not move.isdigit():
            error = "This is not a digit\n"
        elif int(move) not in range(1,10):
            error = "Out of range\n"
        elif board[int(move)] in ('X','O'):
            error = "Cell occupado\n"
        else:
            break
            
        #User entered an invalid choice so display the board again
        display_board(board)
        
    #Input passed validation, parse to int and assign board cell
    board[int(move)] = 'X'
    
    return board

def ai_move(board):
    
    #Generate a random number corresponding to a cell
    move = random.randrange(1,10)
    
    #check board if cell is occupied, if so regenerate
    while board[move] in ('X','O'):
        move = random.randrange(1,10)
    
    #Random number passed, assign to board cell
    board[move] = 'O'
    
    return board

def clear_board():
    #Clear the display and reset the board
    clear_output()
    return ['-',' ',' ',' ',' ',' ',' ',' ',' ',' ']

def coin_toss():
    return random.randint(1,2)
```

```python
#Generate initial board, and possible answers to playing
board = clear_board()
answer = ("y".upper(),"yes".upper())

print("Greetings professor Falken. Shall we play a game of tic-tac-toe.")
while True:
    
    #Determine who goes first
    turn = coin_toss()
    print("Coin Toss...")
    print("You go first." if turn == 1 else "I'll goes first.")
    
    #Ask if ready, anything other than yes will be a no
    if input("Ready (y,yes|n,no):").upper() in answer:
        while display_board(board):
            if turn == 1:
                board = user_move(board)
                turn = 2
                continue
            else:
                board = ai_move(board)
                turn = 1
                continue
                
    #Ask to play again, anything other than yes will be a no
    if input("Play again (y,yes|no,n):").upper() in answer:
        board = clear_board()
        continue
    else:
        break
        
```
